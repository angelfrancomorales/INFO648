```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#sklearn-basic machine learning library in python
from sklearn.model_selection import train_test_split #split data for us
from sklearn.compose import ColumnTransformer #help us with onehot encoding
from sklearn.preprocessing import OneHotEncoder #ovy
from sklearn.linear_model import LinearRegression #our model
from sklearn.pipeline import Pipeline #our model becomes more than regression-workflow
from sklearn.metrics import mean_absolute_error, mean_squared_error #metrics
```

**Part 1: Load and Explore**

```python
df=pd.read_csv('saas_deals.csv')
```

Company Employees, Sales Calls, and Pipeline Days are **numeric**. Industry, Plan Tier, and Region are **categorical**. Our **target** is ACV USD.

**Part 2: Train/Test Split**

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['acv_usd'])
y = df['acv_usd']
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=0,
)
```

```python
X.head()
```

This test will help us estimate how the model will run on new, unseen data. Once the model sees the data again, the metrics would be optimistic.

**Part 3: Build the Pipline**

```python
numeric_features=['company_employees','sales_calls', 'pipeline_days']
categorical_features=['industry','plan_tier', 'region']
target='acv_usd'
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features),
        ('num', 'passthrough', numeric_features)
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression

pipeline = Pipeline(steps=[
    ('prep', preprocessor),
    ('lr', LinearRegression())
])

print(pipeline)
```

**Part 4: Fit and Evaluate**

```python
pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)
```

```python
y_pred
```

```python
y_pred=pipeline.predict(X_test)
preview=pd.DataFrame({'Actual':y_test,'Predict':y_pred})
preview
```

```python
pipeline.fit(X_train,y_train)
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(f'MAE=${mae} and RMSE=${rmse}')
```

Our model's ACV prediction is off by approximately 10,626 per deal, so if the model forecasts a deal at 75,000, the actual value could be anywhere from around 64,000 to 86,000.

**Part 5: Interpret the Coefficients**

```python
import numpy as np
import pandas as pd

# Pull feature names out of the preprocessor and pair them with coefficients
feature_names = pipeline.named_steps['prep'].get_feature_names_out()
coefs = pipeline.named_steps['lr'].coef_
intercept = pipeline.named_steps['lr'].intercept_

coef_table = pd.DataFrame({
    'feature': feature_names,
    'coefficient_$K': coefs.round(2),
})
print(f'intercept = ${intercept:,.0f}')
coef_table
```

An additional Sales call is associated with about 667 more in ACV, holding everything else constant.

Industry tech is associated with about 19,276 less in ACV. These deals are associated with about 19,277 less in ACV compared to Fiance deals (baseline), holding everything else constant.

A Pipeline day is associated with about 47 more in ACV, holding everything else constant.
