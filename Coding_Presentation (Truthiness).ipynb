## **What is Truthiness in Python?**

Truthiness is how Python decides whether a value behaves like **True** or **False** in an **if** statement.

In Python, values do not have to literally be **True** or **False** to be used in conditions.

```python
# False - the boolean false
# None -  the absence of a value (like a null/missing)
# 0 and 0.0 - zero as an integer and a float
# "" an empty string
# [] an empty list
# {} an empty dictionary
# () an empty tuple
# set() an empty set
```

Empty or nothing = falsy

```python
# True - the boolean true
# 1  and -1 - any none-zero number, postive or negative
# "hello" - any non-empty string
# [1, 2, 3] - any non-empty list
# {"name": "Angel"} - any non-empty dictionary
```

Something in it = truthy

**Simple example:**

```python
name = ""

if name:
    print("Name was entered")
else:
    print("Name is missing")
```

#  Examples of Truthiness in Python `if` statements

**Example 1: Checking if a list has data**

```python
sales = []

if sales:
    print("We have sales data")
else:
    print("No sales data available")
```

An empty list is falsy. A list with at least one item is truthy.

```python
sales = [100, 250, 400]

if sales:
    print("We have sales data")
```

Before analyzing a dataset, you may check whether data was loaded successfully.

**Example 2: Zero is falsy, but zero may be valid data**

```python
discount = 0

if discount:
    print("Discount applied")
else:
    print("No discount")
```

This prints **No discount**, but analytically, 0 might have a real value, not missing data.

```python
if discount is not None:
    print("Discount value exists")
else:
    print("Discount is missing")
```

This is one of the biggest analytics mistakes. A value of 0 can mean something real, like zero sales, zero defects, or zero discount. It does not always mean missing.

**Example 3: Empty strings in survey or customer data**

```python
customer_feedback = ""

if customer_feedback:
    print("Feedback received")
else:
    print("No feedback provided")
```

This works when an empty string means the person left the answer blank.

But if the value is "0" as text:

```python
rating = "0"

if rating:
    print("Rating exists")
```

Because "0" is a non-empty string, so it is truthy.

CSV files often read numbers as text. "0" and 0 behave differently in conditions.

**Example 4: Pandas-style thinking**

```python
import pandas as pd

df = pd.DataFrame({
    "customer": ["A", "B", "C"],
    "sales": [100, 0, None]
})

print(df)
```

Bad logic:

```python
for sale in df["sales"]:
    if sale:
        print("Valid sale:", sale)
    else:
        print("Missing or zero:", sale)
```

Problem: this treats 0 and missing values similarly.

Better:

```python
for sale in df["sales"]:
    if pd.notna(sale):
        print("Sale value exists:", sale)
    else:
        print("Missing sale:", sale)
```

In analytics, we usually need to separate **missing values** from **real zero values.**

# **Common Mistakes with Truthiness in Python**

**Mistake 1: Treating 0 as missing**

Bad:

```python
if revenue:
    print("Revenue exists")
```

Better:

```python
revenue = 0
if revenue is not None:
    print("Revenue exists")
```

Why it matters? Revenue of 0 may be meaningful.

**Mistake 2: Confusing empty strings and missing values**

Bad:

```python
customer_name = ""
if customer_name:
    print("Customer name exists")
```

This only catches empty strings, not necessarily all missing values.

Better in Pandas:

```python
import pandas as pd

df = pd.DataFrame({
    "customer_name": ["Alice", "", "Bob", None, "Charlie", "Dora"]
})

print(df["customer_name"].isna())
```

```python
df["customer_name"].replace("", pd.NA)
```

Why it matters? Missing data can appear as None, NaN, or "".

**Mistake 3: Assuming "False" is false**

```python
value = "False"

if value:
    print("This is truthy")
```

Because "False" is a non-empty string.

Better:

```python
value = value == "True"
```

Why it matters? Data from CSVs, forms, or APIs often stores booleans as text.

**Mistake 4: Using truthiness when you need an exact condition**

Bad:

```python
score = 0
if score:
    print("Score exists")
```

Better:

```python
if score is not None:
    print("Score exists")
```

Or for a rule:

```python
if score >= 70:
    print("Passing")
```

Why it matters? Analytics code should be clear about the business rule.

# **Key Takeaways**

Truthiness is Python’s shortcut for deciding whether a value behaves like true or false.

1. Empty values like "", [], {}, None, and 0 are falsy.

2. Most non-empty or non-zero values are truthy.

3. Truthiness makes code shorter, but it can create analytics mistakes.

4. In data analytics, we should always ask:

Am I checking whether something exists, whether it is missing, or whether it meets a specific condition?

Truthiness is helpful, but in analytics, clarity is more important than clever code.
