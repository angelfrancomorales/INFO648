```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/JWarmenhoven/ISLR-python/master/Notebooks/Data/Wage.csv"
wage_df = pd.read_csv(url)

wage_df.head()
wage_df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=wage_df, x='age', y='wage', alpha=0.3)
plt.title('Wage vs Age')
plt.show()
```

Wages tend to increase around the 30 year olds and decrease after the 60 year olds. Earnings ranges between $50 to $200 per hour with a few outliers around the $250 and $300 per hour mark.

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=wage_df, x='year', y='wage')
plt.title('Wage by Year')
plt.show()
```

Here we can observe a slight increase in wages from 2003 to 2009. This means there is an increase in wages through time; however, it is not fast or drastic.

```python
plt.figure(figsize=(8,5))
order = ['1. < HS Grad', '2. HS Grad', '3. Some College', '4. College Grad', '5. Advanced Degree']
sns.boxplot(data=wage_df, x='education', y='wage', order=order)
plt.xticks(rotation=20)
plt.title('Wage by Education Level')
plt.show()
```

Here we can see a similiar averega close to 150 wage for HS Grad and people with some college. There is a significant difference between college graduates and those with advanced degrees, as the latter are closer to or above the 250 wage mark.

We do need to highlight some outliers in the HS Grad and some college that out earn the college grad and advance degree.
